Select a random subset of artists and make a pandas dataframe of all their auction lots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Image
%matplotlib inline
N_ARTISTS = 2000

#grab a few artists
artists = np.random.choice(c.distinct('artistID'), N_ARTISTS)
#make MongoDB cursor
cursor = c.find({'artistID':{'$in':list(artists)}})
df = pd.DataFrame(list(cursor))
del df['_id']

print 'Loaded %d lots belonging to %d artists.' % (df.shape[0], N_ARTISTS)
```

```python
def artist_plot(artist):
    fig, ax = plt.subplots(1,2)
    artist_price_trend = df[df['artistName']==artist][['auctionDate','priceUSD']]
    artist_price_trend['auctionDate'] = pd.to_datetime(artist_price_trend['auctionDate'])
    artist_price_trend.index = artist_price_trend['auctionDate']
    ix = np.unravel_index(0, ax.shape)
    artist_price_trend.ix['2002':].plot(ax=ax[ix], figsize=(12,4))
    #pd.ewma(artist_price_trend.priceUSD.ix['2002':],span=50).plot(style='--', color='red')
    #plt.subplot(len(top_artists),2,i+1)
    #plt.subplot(1,2,2)
    fig1 = artist_price_trend.priceUSD.hist(bins = 25, orientation='horizontal',figsize=(12,4))
    fig1.axes.set_yticks(ticks = [])
    plt.suptitle('%s auction prices over time' % artist,fontsize=20)
```

```python
def artists_works_plot(artist):
    art_price_trend = df[df.artistName == artist][['date','priceUSD','artTitle','images','auctionDate']]
    art_price_trend = art_price_trend[art_price_trend.date != u'']
    art_price_trend.date = art_price_trend.date.str.findall('\d{4}').str.get(0)
    art_price_trend.date = art_price_trend.date.astype('int64')
    
    top_pieces = art_price_trend.sort('priceUSD',ascending=False).iloc[:3]
    
    try:
        top_pieces.images = top_pieces.images.apply(pd.Series)[0].apply(pd.Series).path
    except KeyError:
        print 'No images available'
    #plt.subplot(1, 1 + top_pieces.dropna().shape[0], 1)
    
    plt.scatter(art_price_trend.date, art_price_trend.priceUSD)
    for label, x, y in zip(top_pieces.artTitle, top_pieces.date, top_pieces.priceUSD):
        plt.annotate(label, xy = (x,y), ha='center')
    plt.ylabel('priceUSD')
    plt.xlabel('year')
    plt.title('%s art prices by composition year' % artist)
    return top_pieces
    
```
